### Find Total Impedance
**i**. Add the three impedances.

**ii**. Print `real part` (resistive) and `imaginary part` (reactive).

**iii**. Print `magnitude` and `phase angle` (use abs() and cmath.phase()).

___

### Q.1) Add the three impedances.

```python
Z1 = complex(3, 4)  # Z1 = 3 + 4j
Z2 = complex(2, -3)  # Z2 = 2 - 3j
Z3 = complex(1, 2)  # Z3 = 1 + 2j

# 1. Add the three impedances
Z_total = Z1 + Z2 + Z3
print(f"TOTAL_IMPEDNCES: {Z_total}")
```

___

### Q.2) Print `real part` (resistive) and `imaginary part` (reactive).

```python
print("Real part (Resistive):", Z_total.real)
print("Imaginary part (Reactive):", Z_total.imag)
```

___

### Q.3) Print `magnitude` and `phase angle` (use abs() and cmath.phase()).

```python
# for pahse angle  we use import cmath
import cmath
```

```python
magnitude = abs(Z_total)
phase_angle = cmath.phase(Z_total)

print("Magnitude of Total Impedance:", magnitude)
print("Phase Angle of Total Impedance:", phase_angle, "radians")
```
