sum: sum() adds up numbers in a list.
1.	The power consumed by individual components is listed in a list. Find the total power consumed by adding readings.
2.	A list contains the values of a series of connected resistances. Find its equivalent resistance using the sum.

```python
# 1.	The power consumed by individual components is listed in a list. Find the total power consumed by adding readings.
power_readings = [50, 100, 75, 125]

total_power = sum(power_readings)

print("Total Power Consumed:", total_power, "W")
```

```python
# 2.	A list contains the values of a series of connected resistances. Find its equivalent resistance using the sum.
resistances = [10, 20, 30, 40]

# Use sum() to find total equivalent resistance
total_resistance = sum(resistances)

print("Equivalent Series Resistance:", total_resistance, "Œ©")
```
