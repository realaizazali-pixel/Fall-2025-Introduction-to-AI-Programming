## 4.	Resistance, Voltage, and Current Lists:
You are building a program to analyze five resistor values in a circuit.
1.	Function 1: resistance():  return a list of 5 resistances.
2.	Function 2: voltage():  return voltages across each resistor.
3.	Function 3: current():  take both lists (Resistance & Voltage) and calculate currents.
4.	Display all lists nicely in the main program.
5.	Try two versions:

    	`Return as a tuple.`
        
        `Return as a dictionary with keys like "current_1".`


```python
# Function 1: Resistances
def resistance():
    return [10, 20, 30, 40, 50]
```

```python
# Function 2: Voltages
def voltage():
    return [5, 10, 15, 20, 25]
```

```python
# Function 3: Currents
def current(R, V):
    return tuple(v / r for v, r in zip(V, R))# zip(V, R)  Combines two lists element by element.
```

```python
#  Main Program 
res_list = resistance()
vol_list = voltage()
cur_tuple = current(res_list, vol_list)
```

```python
print("Resistances (Œ©):", res_list)
print("Voltages (V):   ", vol_list)
print("Currents (A):   ", cur_tuple)
```

```python
# Function 3 (dictionary version)
def current_dict(R, V) -> dict[str, float]:
    return {f"current_{i+1}": v / r for i, (v, r) in enumerate(zip(V, R))} # enumerate(...) gives the index (i) so you can label each current
```

```python
# Main Program (dictionary version)
cur_dict = current_dict(res_list, vol_list)

print("\n--- Dictionary Version ---")
print("Resistances (Œ©):", res_list)
print("Voltages (V):   ", vol_list)
print("Currents (A):   ", cur_dict)
```

## 6.	Parallel Resistance:
Now, calculate parallel resistance.
1.	Write a function parallel() that takes a list or tuple of resistances.
2.	Return the equivalent resistance.
3.	Add docstrings and annotations.
4.	Print the function‚Äôs `__annotations__.`


```python
def parallel(resistances):
        # 1: empty input
    if not resistances:
        return 0.0

    # 2: short-circuit check
    if any(r == 0 for r in resistances):
        return 0.0

    # 3: compute reciprocal sum
    reciprocal_sum = 0.0
    for r in resistances:
        reciprocal_sum += 1.0 / r

    # 4: final result (handle zero reciprocal_sum)
    return 1.0 / reciprocal_sum if reciprocal_sum != 0.0 else float('inf')
```

```python
def parallel(resistances: list[float] | tuple[float, ...]) -> float:
    """
    Calculate equivalent resistance for resistors in parallel.

    Parameters:
        resistances (list[float] | tuple[float, ...]): resistor values in ohms

    Returns:
        float: equivalent resistance in ohms.
               - returns 0.0 if list is empty or any resistor == 0 (short).
               - returns float('inf') if reciprocal sum is exactly zero.
    """
print(parallel.__annotations__)
```

```python
# quick examples
print("parallel([10, 20, 30]) ->", parallel([10, 20, 30]), "ohms")
print("parallel([]) ->", parallel([]), "ohms")
print("parallel([0, 10]) ->", parallel([0, 10]), "ohms")
```

```python
def series(*args: float) -> float:
    """
    Calculate total resistance for resistors connected in series.

    Parameters:
        *args (float): Any number of resistor values in ohms.

    Returns:
        float: The total resistance in ohms (0.0 if no values are given).
    """
    return float(sum(args)) if args else 0.0


# Print function annotations
print(series.__annotations__)

```

## 8.	Parallel Resistance with *args:
Do the same for parallel resistance.
1.	Define parallel(*args) using tuple packing.
2.	Pass any number of resistances.
3.	Return the equivalent resistance.
4.	Add docstrings and annotations.
5.	Print `__annotations__.`


```python
def parallel(*args: float) -> float:
    """
    Calculate equivalent resistance for resistors connected in parallel.

    Parameters:
        *args (float): Any number of resistor values in ohms.

    Returns:
        float: Equivalent resistance in ohms.
               - 0.0 if no resistors are provided.
               - 0.0 if any resistor is 0 (short circuit).
               - float('inf') if reciprocal sum = 0.
    """
       # No resistors given
    if not args:
        return 0.0
    
    # Short-circuit case
    if any(r == 0 for r in args):
        return 0.0
    
    # Compute reciprocal sum
    reciprocal_sum = sum(1.0 / r for r in args)
    
    # Return equivalent resistance
    return 1.0 / reciprocal_sum if reciprocal_sum != 0 else float('inf')
```

```python
# Print function annotations
print(parallel.__annotations__)
```

```python
# Example usage
print("parallel(10, 20, 30) ->", parallel(10, 20, 30), "ohms")
print("parallel(5.0, 5.0) ->", parallel(5.0, 5.0), "ohms")
print("parallel() ->", parallel(), "ohms")
print("parallel(0, 10) ->", parallel(0, 10), "ohms")
```

## 13.	Chain Rule 


```python
import math

def y(x):
    return x**2

def z(y_val):
    return math.sin(y_val)
```

```python
def dz_dx(x):
    return math.cos(x**2) * (2*x)
```

```python
import torch

# Function to compute derivative using autograd
def dz_dx_autograd(x_value):
    x = torch.tensor(x_value, requires_grad=True)
    y = x**2
    z = torch.sin(y)
    z.backward()         # compute gradient
    return x.grad.item() # return derivative
```

```python
x_values = [0.5, 1.0, 2.0]

print("x\tManual Derivative\tAutograd Derivative")
for val in x_values:
    manual = dz_dx(val)
    auto = dz_dx_autograd(val)
    print(f"{val}\t{manual:.6f}\t\t{auto:.6f}")
```
