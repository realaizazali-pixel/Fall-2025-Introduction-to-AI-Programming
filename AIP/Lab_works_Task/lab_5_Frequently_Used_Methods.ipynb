### enumerate:
 enumerate () adds numbers (index) to items in a list or other collection.
1.	A list contains voltage readings from 5 measurement points. Use enumerate () to display the reading number with each voltage.
2.	A student records resistance values of resistors in a lab. Use enumerate () to print the resistor number and its value.


```python
# 1.	A list contains voltage readings from 5 measurement points. Use enumerate () to display the reading number with each voltage.
Voltage_reading=[20,30,40,50,60]
for i,v in enumerate(Voltage_reading):
    print(i,v)
```

```python
# 2.	A student records resistance values of resistors in a lab. Use enumerate () to print the resistor number and its value.
student_record=[3,4,5,6,7,8,9]
for i,v in enumerate(student_record):
    print(f"R {i}, {v}ohm")
```

### zip:
 zip () combines two or more lists element by element.

1.	To lists, one list has current readings, and another has voltage readings. Use zip() to print the pair of current and voltage.
2.	Combine a list of component names with their ratings (resistor with ohms, capacitor with ¬µF).

### map:
 map() applies a function to every item in a list.

`Note:` lambda is a short way to write a function in one line. Example: lambda x: x*2 means a function that doubles x.
1.	Convert a list of temperatures from Celsius to Fahrenheit using the formula F = (C √ó 9/5) + 32.
2.	Convert a list of voltages in volts to kilovolts using V_kV = V / 1000.

```python
# 1.	Convert a list of temperatures from Celsius to Fahrenheit using the formula F = (C √ó 9/5) + 32.
tempature_in_C=[2,3,4,5,6]
```

```python
temp_in_F= list(map(lambda c:(c*(9/5)+32),tempature_in_C))
print(f"tempature in Faranheit: {temp_in_F}")
```

```python
# 2.	Convert a list of voltages in volts to kilovolts using V_kV = V / 1000.
vol=[23,24,25,26,27]
```

```python
v_KV=list(map(lambda v:(v/1000),vol))
print(f'voltage in killo watt: {v_KV}')
```

### abs:

```python
# 1.	The function generator values are stored in a list over a cycle. Find the absolute value of alternating current readings.

generator_readings = [2.5, -2.4, 2.3, -2.2, 2.1, -2.0]

# Use abs() to convert all readings to positive (magnitude)
magnitude_readings = [abs(i) for i in generator_readings]

print("generator readings (absolute values):", magnitude_readings)
```

### divmod: 
divmod() returns the quotient and remainder of the division.
1.	87 optical fiber cables are drawn to provide high-speed internet for a university campus. The duct can contain 10 cables at most. Divide the total wires into bundles and find the number of completed filed ducts and how many more cables can be placed in the final duct.
2.	Each group brings 42 LEDS for the DLD project. Divide the total number of LEDs into groups of 12, and find the groups and extra LEDs.


```python
# 1.	87 optical fiber cables are drawn to provide high-speed internet for a university campus. The duct can contain 10 cables at most. Divide the total wires into bundles and find the number of completed filed ducts and how many more cables can be placed in the final duct.

total_cables = 87
duct_capacity = 10

# divmod returns (quotient, remainder)
full_ducts, remaining_cables = divmod(total_cables, duct_capacity)

print(f"Completed full ducts: {full_ducts}")
print(f"Cables that can still be placed in the last duct: {remaining_cables}")
```

```python
# 2.	Each group brings 42 LEDS for the DLD project. Divide the total number of LEDs into groups of 12, and find the groups and extra LEDs.
total_leds = 42
group_size = 12

groups, extra_leds = divmod(total_leds, group_size)

print(f"Full groups of LEDs: {groups}")
print(f"Extra LEDs remaining: {extra_leds}")
```
