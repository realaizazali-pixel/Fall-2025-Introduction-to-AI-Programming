## 1.	Task: Voltage Level Classification 
1.	A text file contains circuit voltage readings (one per line).
2.	Task: Read the file and classify each voltage as:
	1. Low (< 110V)
	2. Medium (110V‚Äì220V)
	3. High (> 220V)
3.	Save the classification results into a new file.


```python
with open("voltage.txt", "r") as file:
    readings = file.readlines()
    print(readings)
```

```python
results = []  # store classified results

for line in readings:

    voltage = float(line.strip())  # convert string to float
    if voltage < 110:Level = "Low"
    elif 110 <= voltage <= 220:Level = "Medium"
    else:Level= "High"
    results.append((voltage, Level))
results
```

```python
with open("classified_voltages.txt", "w") as f:
    for voltage,Level in results:
        f.write(f"{voltage}V , {Level}\n")
```

## 4.	Task: File Extension Validator (General Engineering Reports)
1.	A directory has multiple engineering reports.
2.	Task: Write a program to scan all files.
    1.	If file ends with .pdf ‚Üí Valid Report
    2.	If file ends with .docx ‚Üí Needs Conversion
    3.	Otherwise ‚Üí Invalid Format
3.	Count and display the numbers of each type.


```python
import os

def validate_reports(directory):
    # Counters
    valid_count = 0
    convert_count = 0
    invalid_count = 0

    # Scan all files in the give path
    for file in os.listdir(directory):
        if file.endswith(".pdf"):
            print(f"{file} : Valid Report")
            valid_count += 1
        elif file.endswith(".docx"):
            print(f"{file} : Needs Conversion")
            convert_count += 1
        else:
            print(f"{file} : Invalid Format")
            invalid_count += 1

    # Summary
    print("\nSummary:")
    print(f"Valid Reports (.pdf): {valid_count}")
    print(f"Needs Conversion (.docx): {convert_count}")
    print(f"Invalid Format (others): {invalid_count}")


# Example usage
directory_path = "C:/EngineeringReports"  # <-- change to your folder path
validate_reports(r"D:\Electrical Engineering Computing & AI\3rd Semester\Introduction to AI Programming\AI LAB")
```
